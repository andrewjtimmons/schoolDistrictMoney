```python
import numpy as np
import pandas as pd
```

```python
money_df = pd.read_csv("census_money_data/elsec13t.txt")
diversity_df = pd.read_csv("nces_diversity_data/elsi.csv")
```

```python
#there are some strange items in these DFs like = signs, quotes and crosses.
#let's remove them.

def convert_funky_stuff_to_nan(column):
    # This line is for values that seem valid but have junk characters 
    # around them that should never be there.
    column.replace(r"[=\"†]+",np.nan, regex = True, inplace = True)
    # This line or lines is for values that might be valid in one context, like -40.43534 for a latitiude
    # but when they are the only value in a cell they are not valid.  
    column.replace("-",np.nan, inplace = True)
    column.replace("n/a",np.nan, inplace = True)
    

for column in diversity_df:
    convert_funky_stuff_to_nan(diversity_df[column])

for column in money_df:
    convert_funky_stuff_to_nan(money_df[column])
```

```python
# Sort by id to merge on, see if there are any np.nan values, drop them, and inspect them
diversity_df.sort(["Agency ID - NCES Assigned [District] Latest available year"], inplace = True)
diversity_df["Agency ID - NCES Assigned [District] Latest available year"].dropna(inplace = True)
diversity_df["Agency ID - NCES Assigned [District] Latest available year"]
```

```python
# Sort by id to merge on, see if there are any np.nan values, drop them, and inspect them
money_df.sort(["NCESID"], inplace = True)
money_df["NCESID"].dropna(inplace = True)
money_df["NCESID"]

```

```python
#join the DFs together on the NCES id number
money_diversity_df = pd.merge(money_df, diversity_df,  left_on = "NCESID", right_on = "Agency ID - NCES Assigned [District] Latest available year")
```

```python
#there are some schools that have nan total students. Let's drop those.
money_diversity_df.dropna(inplace = True, subset = ['Total Students (UG PK-12) [District] 2012-13','Total Students [Public School] 2012-13'])
```

```python
ids = ['NCESID', 'PPCSTOT']
columns_to_examine = ['NCESID', 'PPCSTOT', 
                      'Hawaiian Nat./Pacific Isl. Students [Public School] 2012-13', 
                      'American Indian/Alaska Native Students [Public School] 2012-13']
new_df = money_diversity_df[columns_to_examine]
new_df.fillna(0, inplace=True)
new_df = new_df.astype(int)

```

```python
def expand_students(d):
    """Takes a dataframe with students that are aggregated and expands them into one row per students in a series"""
    return pd.Series(d['Hawaiian Nat./Pacific Isl. Students [Public School] 2012-13'].values*['Hawaiian Nat./Pacific Isl. Student']+ 
                     d['American Indian/Alaska Native Students [Public School] 2012-13'].values*['American Indian/Alaska Native Student'])

new_df = new_df.groupby(ids).apply(expand_students).reset_index()
```
