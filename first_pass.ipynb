#Clean the data and join our data sets

```python
#there are some strange items in these DFs like = signs, quotes and crosses.
#let's remove them.

def convert_funky_stuff_to_nan(column):
    # This line is for values that seem valid but have junk characters 
    # around them that should never be there.
    column.replace(r"[=\"†]+",np.nan, regex = True, inplace = True)
    # This line or lines is for values that might be valid in one context, like -40.43534 for a latitiude
    # but when they are the only value in a cell they are not valid.  
    column.replace("-",np.nan, inplace = True)
    column.replace("n/a",np.nan, inplace = True)
    

for column in diversity_df:
    convert_funky_stuff_to_nan(diversity_df[column])

for column in money_df:
    convert_funky_stuff_to_nan(money_df[column])
```

```python
# Sort by id to merge on, see if there are any np.nan values, drop them, and inspect them
diversity_df.sort(["Agency ID - NCES Assigned [District] Latest available year"], inplace = True)
diversity_df["Agency ID - NCES Assigned [District] Latest available year"].dropna(inplace = True)
diversity_df["Agency ID - NCES Assigned [District] Latest available year"]
```

# Functions for the dataframes

```python
def expand_students(df, **kwargs):
    """
    kwargs takes group_labels and student_labels
    Takes a dataframe with students that are aggregated and expands them into one row per students in a series
    """
    list_student_rows = []
    for group_label, student_label in zip(kwargs['group_labels'], kwargs['student_labels']):
        for value in df[group_label].values:
            for x in range(0, value):
                list_student_rows.extend([student_label])
        #list_student_rows.extend(d[group_label].values*[student_label])
    return pd.Series(list_student_rows)
```

```python
def remove_outliers(df, column_name):
    return df[np.abs(df[column_name] - df[column_name].mean())<=(3*df[column_name].std())]
```

```python
def get_median_PPCSTOT_for_labels(df, student_labels):
    # generates the median PPCSTOT for a set of labels
    labels_with_medians = []
    for student_label in student_labels:
        median_per_pupil_spending = df['PPCSTOT'][df['student_label'] == student_label].median()
        labels_with_medians.append((student_label, median_per_pupil_spending))
    labels_with_medians.append(('all student', df['PPCSTOT'].median()))
    return labels_with_medians

def get_average_PPCSTOT_for_labels(df, student_labels):
    labels_with_averages = []
    for student_label in student_labels:
        average_per_pupil_spending = df['PPCSTOT'][df['student_label'] == student_label].mean()
        labels_with_averages.append((student_label, average_per_pupil_spending))
    labels_with_averages.append(('all student', df['PPCSTOT'].mean()))
    return labels_with_averages

def pretty_print_labels_with_statistic(labels_with_statistics, statistic_type):
    """takes in a list and a type of stat and prints it nice."""   
    for label, statistic in labels_with_statistics:
        print "%s spending per pupil for %ss is %s" % (statistic_type, label, statistic)
```

```python
# expand the df and clean it
expanded_df = expanded_df.groupby(ids).apply(expand_students, group_labels = group_labels, student_labels = student_labels).reset_index()
del expanded_df['level_2']
expanded_df.rename(columns={0: 'student_label'}, inplace=True)
```

```python
#test functions to make sure it is making as many rows as the original dataframe
expected_sum = 0
for group_label in group_labels:
    expected_sum += money_diversity_df[group_label].sum()
    
expected_sum = int(expected_sum)
assert expanded_df.shape[0] == expected_sum, "expanded_df shape is %s, expected sum is %s" % (expanded_df.shape[0], expected_sum)
```

```python
#test functions to make sure it is making as many rows as the original dataframe for each label
for group_label, student_label in zip(group_labels, student_labels):
    money_diversity_student_count = money_diversity_df[group_label].sum()
    expanded_student_count = expanded_df[expanded_df['student_label'] == student_label].shape[0]
    assert expanded_student_count == money_diversity_student_count, "expanded_df count is %s, money_diversity_df count is %s for label %s" % (expanded_student_count, money_diversity_student_count, student_label)
     
```

```python
# lets explore the amount given to students by race labels
```

```python
#remove outliers
expanded_df = remove_outliers(expanded_df, 'PPCSTOT')
```

```python
labels_with_median = get_median_PPCSTOT_for_labels(expanded_df,student_labels)
pretty_print_labels_with_statistic(labels_with_median, "median")
```

```python
labels_with_averages = get_average_PPCSTOT_for_labels(expanded_df, student_labels)
pretty_print_labels_with_statistic(labels_with_averages, "average")
```
